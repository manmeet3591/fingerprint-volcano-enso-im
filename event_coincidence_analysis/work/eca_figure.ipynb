```python
import numpy as np
import matplotlib.pyplot as plt
import random
import warnings
from scipy import stats
import datetime as dt
```

```python
prec_rate_mann_2009 = np.genfromtxt('prec_rate_mann_2009.csv', delimiter=",", dtype=float)
p_prec_rate_mann_2009 = np.genfromtxt('p_prec_rate_mann_2009.csv', delimiter=",", dtype=float)

prec_rate_moy_2002 = np.genfromtxt('prec_rate_moy_2002.csv', delimiter=",", dtype=float)
p_prec_rate_moy_2002 = np.genfromtxt('p_prec_rate_moy_2002.csv', delimiter=",", dtype=float)

prec_rate_li_2013 = np.genfromtxt('prec_rate_li_2013.csv', delimiter=",", dtype=float)
p_prec_rate_li_2013 = np.genfromtxt('p_prec_rate_li_2013.csv', delimiter=",", dtype=float)

prec_rate_kobb_2003 = np.genfromtxt('prec_rate_kobb_2003.csv', delimiter=",", dtype=float)
p_prec_rate_kobb_2003 = np.genfromtxt('p_prec_rate_kobb_2003.csv', delimiter=",", dtype=float)


t = np.arange(1,11)
idx1_mann_2009 = p_prec_rate_mann_2009<0.05
idx2_mann_2009 = p_prec_rate_mann_2009<0.1

idx1_moy_2002 = p_prec_rate_moy_2002<0.05
idx2_moy_2002 = p_prec_rate_moy_2002<0.1

idx1_li_2013 = p_prec_rate_li_2013<0.05
idx2_li_2013 = p_prec_rate_li_2013<0.1

idx1_kobb_2003 = p_prec_rate_kobb_2003<0.05
idx2_kobb_2003 = p_prec_rate_kobb_2003<0.1
```

```python
plt.scatter(t,prec_rate_moy_2002,label='Moy et. al 2002', facecolors='none',  edgecolors='darkmagenta', s=100)
plt.scatter(t,prec_rate_mann_2009,label='Mann et. al 2009', facecolors='none',  edgecolors='b', s=100)
plt.scatter(t,prec_rate_li_2013,label='Li et. al 2013', facecolors='none',  edgecolors='red', s=100)
#plt.scatter(t,prec_rate_kobb_2003,label='Cobb et. al 2003', facecolors='none',  edgecolors='black', s=100)

plt.scatter(t[idx1_mann_2009],prec_rate_mann_2009[idx1_mann_2009], label='sign at 95%', marker="+", s=50, color='darkorange')
plt.scatter(t[idx2_mann_2009],prec_rate_mann_2009[idx2_mann_2009], label='sign at 90%', marker="x", s=50, color='lightgreen')


plt.scatter(t[idx1_moy_2002],prec_rate_moy_2002[idx1_moy_2002], marker="+", s=50, color='darkorange')
plt.scatter(t[idx2_moy_2002],prec_rate_moy_2002[idx2_moy_2002], marker="x", s=50, color='lightgreen')

plt.scatter(t[idx1_li_2013],prec_rate_li_2013[idx1_li_2013], marker="+", s=50, color='darkorange')
plt.scatter(t[idx2_li_2013],prec_rate_li_2013[idx2_li_2013], marker="x", s=50, color='lightgreen')

#plt.scatter(t[idx1_kobb_2003],prec_rate_kobb_2003[idx1_kobb_2003], marker="+", s=50, color='darkorange')
#plt.scatter(t[idx2_kobb_2003],prec_rate_kobb_2003[idx2_kobb_2003], marker="x", s=50, color='lightgreen')

plt.xlim([0,10])
plt.xlabel('Coincidence interval $Δ$T [years]')
plt.ylabel('Precursor coincidence rate \n (Volcano preceded by El Nino)')
plt.title('ENSO')
plt.legend()
plt.grid()
```
