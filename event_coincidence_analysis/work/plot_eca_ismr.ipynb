```python
import numpy as np
import matplotlib.pyplot as plt
import random
import warnings
from scipy import stats
import datetime as dt
```

```python
prec_rate = np.genfromtxt('prec_rate.csv', delimiter=",", dtype=float)
p_prec_rate = np.genfromtxt('p_prec_rate.csv', delimiter=",", dtype=float)
t = np.arange(1,11)
idx1 = p_prec_rate<0.05
idx2 = p_prec_rate<0.01
idx3 = p_prec_rate<0.1
```

```python
plt.scatter(t,prec_rate,label='Obs rate', facecolors='none',  edgecolors='g', s=100)
plt.scatter(t[idx1],prec_rate[idx1], label='sign at 95%', marker="+", s=50)
#plt.scatter(t[idx2],prec_rate[idx2], label='sign at 99%', marker="x")
plt.scatter(t[idx3],prec_rate[idx3], label='sign at 90%', marker="x", s=50)
plt.xlim([0,10])
plt.xlabel('Lag parameter $τ$ [years]')
#plt.xlabel('Coincidence interval $Δ$T [years]')
plt.ylabel('Precursor coincidence rate \n (Volcano preceded by IM drought)')
plt.title('All SH Volcanic Eruptions')
plt.legend()
plt.grid()
```
