```python
import numpy as np
import matplotlib.pyplot as plt
import random
import warnings
from scipy import stats
import datetime as dt
```

```python
volc = np.genfromtxt('nature14565-s6_sigl2015.csv', delimiter=",", dtype=float)
data_sinha_2015 = np.genfromtxt('ncomms7309-s3_sinha_sahiya_cave_im2015.csv', delimiter=",", dtype=float) # ismr reconstruction
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
start_year = np.trunc(data_sinha_2015[-1,0])
end_year = np.trunc(data_sinha_2015[0,0])
yy_sinha_2015 = np.arange(start_year,end_year)
ismr_sinha_2015 = np.zeros((yy_sinha_2015.shape[0]))

for i in range(yy_sinha_2015.shape[0]):
    x1 = find_nearest(data_sinha_2015[:,0], yy_sinha_2015[i])
    idx = np.asarray(np.where(data_sinha_2015[:,0] == x1))
    x2 = data_sinha_2015[idx-1,0]
    y1 = data_sinha_2015[idx,1]
    y2 = data_sinha_2015[idx-1,1]
    y = y1 + ((y2[0,0]-y1[0,0])/(x2[0,0]-x1))*(yy_sinha_2015[i]-x1)
    ismr_sinha_2015[i] = y
ismr_sinha_2015 = -ismr_sinha_2015
```

```python
plt.plot(data_sinha_2015[:,0],-data_sinha_2015[:,1])
plt.plot(yy_sinha_2015,ismr_sinha_2015)
```

Negative delta 18 O values correspond to positive ISMR

```python
ismr_sinha_norm = np.zeros(ismr_sinha_2015.shape[0])
win = 30 # 30 year climatology
ismr_sinha_norm[0:int(win/2)] = (ismr_sinha_2015[0:int(win/2)] - np.mean(ismr_sinha_2015[0:win]))/np.std(ismr_sinha_2015[0:win])
for i in range(int(win/2),ismr_sinha_2015.shape[0]-int(win/2)):
    ismr_sinha_norm[i] = (ismr_sinha_2015[i] - np.mean(ismr_sinha_2015[i-int(win/2):i+int(win/2)]))/np.std(ismr_sinha_2015[i-int(win/2):i+int(win/2)])
ismr_sinha_norm[ismr_sinha_2015.shape[0]-int(win/2):] = (ismr_sinha_2015[ismr_sinha_2015.shape[0]-int(win/2):] - np.mean(ismr_sinha_2015[ismr_sinha_2015.shape[0]-int(win/2):]))/np.std(ismr_sinha_2015[ismr_sinha_2015.shape[0]-int(win/2):])
plt.plot(yy_sinha_2015, ismr_sinha_norm)
```

```python
np.savetxt('nature14565-s6_sigl2015_v2.csv', ismr_sinha_norm, delimiter=",") 
```

```python
dvolc = np.zeros((ismr_sinha_norm.shape[0]))
dvolc_source = np.zeros((ismr_sinha_norm.shape[0]))
for i in range(ismr_sinha_norm.shape[0]):
    for j in range(volc[:,0].shape[0]):
        if (yy_sinha_2015[i]==volc[j,0]):
            dvolc[i] = volc[j,2]
            dvolc_source[i] = volc[j,1]
print(dvolc_source)
```

```python
# Generate event series for volcanic eruptions and Nino3
im_drought_sinha_2015 = np.zeros((ismr_sinha_norm.shape[0]))

all_volc = np.zeros((ismr_sinha_norm.shape[0]))
strong_volc = np.zeros((ismr_sinha_norm.shape[0]))

all_tropical = np.zeros((ismr_sinha_norm.shape[0]))
strong_tropical = np.zeros((ismr_sinha_norm.shape[0]))

all_nh = np.zeros((ismr_sinha_norm.shape[0]))
strong_nh = np.zeros((ismr_sinha_norm.shape[0]))

all_sh = np.zeros((ismr_sinha_norm.shape[0]))
strong_sh = np.zeros((ismr_sinha_norm.shape[0]))

im_drought_sinha_2015[ismr_sinha_norm<=-1.0] = 1.0

all_volc[dvolc!=0] = 1.0
strong_volc[(np.abs(dvolc)>3.7)] = 1.0

all_tropical[(dvolc_source==1.0)] = 1.0
strong_tropical[(dvolc_source==1.0)&(np.abs(dvolc)>3.7)] = 1.0

all_nh[(dvolc_source==2.0)] = 1.0
strong_nh[(dvolc_source==2.0)&(np.abs(dvolc)>3.7)] = 1.0

all_sh[(dvolc_source==3.0)] = 1.0
strong_sh[(dvolc_source==3.0)&(np.abs(dvolc)>3.7)] = 1.0
```

```python
np.savetxt('im_drought_sinha_2015.txt', im_drought_sinha_2015, delimiter=",")
np.savetxt('all_volc_sinha_2015.txt', all_volc, delimiter=",")
np.savetxt('strong_volc_sinha_2015.txt', strong_volc, delimiter=",")
np.savetxt('all_tropical_sinha_2015.txt', all_tropical, delimiter=",")
np.savetxt('strong_tropical_sinha_2015.txt', strong_tropical, delimiter=",")
np.savetxt('all_nh_sinha_2015.txt', all_nh, delimiter=",")
np.savetxt('strong_nh_sinha_2015.txt', strong_nh, delimiter=",")
np.savetxt('all_sh_sinha_2015.txt', all_sh, delimiter=",")
np.savetxt('strong_sh_sinha_2015.txt', strong_sh, delimiter=",")
```
