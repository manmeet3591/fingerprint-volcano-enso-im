```python
volc = np.genfromtxt('nature14565-s6_sigl2015.csv', delimiter=",", dtype=float)
data_yadava_2004 = np.genfromtxt('ismr_yadava_2004.txt', delimiter=",", dtype=float) # ismr reconstruction
yy_yadava_ann = np.arange(1666,1997)
ismr_yadava_ann = np.zeros((yy_yadava_2004.shape[0]))
xp = data_yadava_2004[::-1,0]
fp = data_yadava_2004[::-1,1]
ismr_yadava_ann = np.interp(yy_yadava_2004, xp, fp)
```

```python
plt.figure(figsize=(16,6))
plt.plot(yy_yadava_ann, ismr_yadava_ann, label='Speleotherm based ISMR reconstruction', alpha=0.8)
plt.legend()
plt.grid()
```

```python
# Singular Spectral Analysis as in Sinha et al 2015
def X_to_TS(X_i):
    """Averages the anti-diagonals of the given elementary matrix, X_i, and returns a time series."""
    # Reverse the column ordering of X_i
    X_rev = X_i[::-1]
    # Full credit to Mark Tolonen at https://stackoverflow.com/a/6313414 for this one:
    return np.array([X_rev.diagonal(i).mean() for i in range(-X_i.shape[0]+1, X_i.shape[1])])

def ssa(F):
    N = F.shape[0]
    L = 30# The window length.
    K = N - L + 1 # The number of columns in the trajectory matrix.
    # Create the trajectory matrix by pulling the relevant subseries of F, and stacking them as columns.
    X = np.column_stack([F[i:i+L] for i in range(0,K)])
    # Note: the i+L above gives us up to i+L-1, as numpy array upper bounds are exclusive. 
    d = np.linalg.matrix_rank(X) # The intrinsic dimensionality of the trajectory space.
    U, Sigma, V = np.linalg.svd(X)
    V = V.T # Note: the SVD routine returns V^T, not V, so I'll tranpose it back here. This may seem pointless, 
    X_elem = np.array( [Sigma[i] * np.outer(U[:,i], V[:,i]) for i in range(0,d)] )
    n = min(12,d) # In case of noiseless time series with d < 12.
    # Convert elementary matrices straight to a time series - no need to construct any Hankel matrices.
    F_i = np.zeros((n,N))
    for i in range(n):
        F_i[i,:] = X_to_TS(X_elem[i])
    return F_i
```

```python
ismr_yadava_ann_ssa = ssa(ismr_yadava_ann)
```

```python
# Trend from Singular Spectral Analysis
plt.figure(figsize=(16,6))
plt.plot(yy_yadava_ann, ismr_yadava_ann, label='delta 18O', alpha=0.8)
plt.plot(yy_yadava_ann, ismr_yadava_ann_ssa[0], label='delta 18O trend', alpha=0.8)
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=(16,6))
plt.plot(yy_yadava_ann, ismr_yadava_ann-ismr_yadava_ann_ssa[0], label='SSA detrended delta 18O', alpha=0.8)
plt.legend()
plt.gca().invert_yaxis()
plt.grid()
```

```python
ismr_yadava_ann_ssa_anom = ismr_yadava_ann-ismr_yadava_ann_ssa[0]
ismr_yadava_ann_ssa_anom_norm  = (ismr_yadava_ann_ssa_anom - np.mean(ismr_yadava_ann_ssa_anom))/(np.std(ismr_yadava_ann_ssa_anom))
plt.figure(figsize=(16,6))
plt.plot(yy_yadava_ann, ismr_yadava_ann_ssa_anom_norm, label='norm delta 18O', alpha=0.8)
plt.legend()
plt.gca().invert_yaxis()
plt.grid()
```
