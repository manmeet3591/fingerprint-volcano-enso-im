```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yy = np.arange(865,1836)
dnino3_jjas = np.genfromtxt('tas_Amon_IPSL-CM5A-LR_past1000_r1i1p1_0865_1835_anom_nino3_jjas_ymean.csv', delimiter=",")
dnino3_djf = np.genfromtxt('tas_Amon_IPSL-CM5A-LR_past1000_r1i1p1_0865_1835_anom_nino3_djf_ymean.csv', delimiter=",")[:971]
dnino3 = np.genfromtxt('tas_Amon_IPSL-CM5A-LR_past1000_r1i1p1_0865_1835_anom_nino3_ymean.csv', delimiter=",")

dimhc = np.genfromtxt('va_Amon_IPSL-CM5A-LR_past1000_r1i1p1_0865_1835_anom_850m200_hc_jjas_ymean.csv', delimiter=",")
dim = np.genfromtxt('pr_Amon_IPSL-CM5A-LR_past1000_r1i1p1_0865_1835_anom_ci_jjas_ymean.csv', delimiter=",")

dstrong1 = np.genfromtxt('strong1.txt', delimiter=",")
```

```python
plt.figure(figsize=(20,10))
plt.rcParams.update({'font.size': 22})
plt.plot(yy,dnino3_jjas)
plt.plot(yy,dnino3_djf)
plt.grid()
```

```python
strong1_en = np.empty(0)
strong1_ln = np.empty(0)
strong1_n = np.empty(0)
for i in range(dstrong1.shape[0]):
    idx = np.where(yy==dstrong1[i])
    if dnino3_djf[idx[0][0]+1]>=0.5:
        #print("first if working")
        strong1_en = np.append(strong1_en, yy[idx[0][0]])
    elif dnino3_djf[idx[0][0]+1]<=-0.5:
        #print("second if working")
        strong1_ln = np.append(strong1_ln, yy[idx[0][0]])
    else:
        strong1_n = np.append(strong1_n, [yy[idx[0][0]]])
        #print("third if working")
```

```python
strong1_en = np.array(strong1_en).astype(int)
strong1_ln = np.array(strong1_ln).astype(int)
strong1_n = np.array(strong1_n).astype(int)

np.savetxt('strong1_en.txt', strong1_en, delimiter=",", fmt="%d") 
np.savetxt('strong1_ln.txt', strong1_ln, delimiter=",", fmt="%d") 
np.savetxt('strong1_n.txt', strong1_n, delimiter=",", fmt="%d") 
```
